# Training models
## Linear Regression 
- Un modelo lineal es la suma ponderada de las caracteristicas input adicionandole una constante llamada bias o sesgo

La ecuación general de un modelo de regresión lineal es:

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n 
$$

Donde:
- $y$ es el valor predicho.
- $\beta_0$ es el intercepto o bias.
- $\beta_1, \beta_2, \dots, \beta_n$  son los coeficientes de las variables independientes.
- $x_1, x_2, \dots, x_n$ son las variables independientes (predictoras).
 
## Normal Equation
- Con esta ecuacion se encuentra el valor de cada uno de los parametros del modelo que minimiza la funcion de coste
$$
\hat{\beta} = (X^T X)^{-1} X^T y
$$

- $\hat{\beta}$ es el valor de $\beta$ que minimiza la función de costo.
- $y$ es el vector de valores objetivo que contiene desde $y^{(1)}$ hasta $y^{(m)}$.


```python
import numpy as np

np.random.seed(42)
m = 100
X = 2 * np.random.rand(m , 1)
y = 4 + 3 * X + np.random.randn(m, 1)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import add_dummy_feature

X_b = add_dummy_feature(X)
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
theta_best
```

```python
X_new = np.array([[0], [2]])
X_new_b = add_dummy_feature(X_new)
y_predict = X_new_b @ theta_best
y_predict
```

```python
plt.plot(X, y, "b.")
plt.plot(X_new, y_predict, "r-", label="Predictions")
plt.xlabel("$x_1$")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axis([0, 2, 0, 14])
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
lin_reg.predict(X_new)
```
