# Capitulo 1
## El panorama del Machine Learning
- ML ya ha estado desde los 90s y ha permitido que un monton de sistemas funcionen adecuadamente como el filtro de spam
## Que es machine learning
- el ml es cuando a una maquina se le da una tarea T con unos datos de entrenamiento E y que medira su performance en H, si despues de la tarea T entrenada con E, H mejora, esta aprendiendo
## Por que utilizr ML
- Aplicar algoritmos de ml en grandes datasets para descubrir relaciones entre los mismos es llamado Data Minning
- Usar ML puede ayudar a resolver tareas complejas que no tienen algoritmos definidos que las resuelvan
- ML es bueno para:
    - descubrir conclusiones en grandes cantidades de datos
    - tareas que requieran una larga lista de condiciones
    - tareas complejas sin algoritmos definidos
    - ambientes que varian constantemente
## Tipos de sistemas ML
- Se pueden clasificar segun varios criterio
    - si son entrenados con o sin supervision humana (supervised, unsupervised, semisupervised, reinforcement learning)
    - si pueden o no aprender de manera increamental mientras estan en marcha (online vs batch learning)
    - si comparan datos nuevos con datos existentes o si detectan patrones en los datos para construir un modelo predictivo (instance-based o model-based learning)
## Supervised / Unsupervised learning
### Supervisado
- en el aprendizaje supervisado es en el que los datos tienen el valor de los datos y tambien su resultado esperado, esto es comun para tareas de clasificacion o de prediccion, para entrenar el sistema es necesario darle muchos ejemplos. 
- Algoritmos de aprendizaje supervisado:
    - K-Nearest Neighbors
    - regresion lineal
    - regresion logistica
    - Support vector machines (SVMs)
    - Decision Trees and Random forests
    - Redes neuronales
### No supervisado
- En el aprendizaje no supervisado, los datos no estan marcados
- Algunos algoritmos de aprendizaje no supervisado
    - Clustering
        - K-means
        - Hierarchical Cluster Analysis (HCA)
        - Expectation Maximization
    - Visualizaton and dimensionality reduction
        - principal component analysis (PCA)
        - Kernel PCA
        - Locally-linear Embedding (LLE)
        - T-distribitet stochastic neighbor embbeding (t-SNE)
    - Association rule learning
        - Apriori 
        - Eclat
- Los algoritmos de clustering  se encargan de agrupar una serie de datos con relaciones comunes
- Los algoritmos de visualizacion brindan una representacion visual de los datos, sirven para entender como los datos estan organizados
- Otra funcion de los algoritmos de aprendizaje no supervisado es dimensionality reduction, en la que los datos son simplificados sin perder mucha informacion, esto se puede hacer uniendo dos caracteristicas relacionadas en una sola lo que se llama Feature extraction
- Es recomendable reducir los training sets usando dimensionality reduction antes de introducirlos a otro algoritmo de ML
- Los algoritmos de aprendizaje no supervisado tambien pueden usarse para la deteccion de anomalias, en la que los algoritmos se entrenan con instancias normales y estos detectan si una nueva instancia es normal o no.
- Otra tarea no supervisada es association rule learning, en la que el algoritmo busca entre grandes cantidades de datos relaciones entre sus atributos.
### Aprendizaje semisupervisado
- En este tipo de aprendizaje los datos estan parcialmente etiquetados, hay unos que estan etiquetados pero lo mas usual es que la gran mayoria esten sin etiquetar 
- la mayoria de estos algoritmos son una combinacion entre algoritmos de aprendizaje no supervisado y supervisado, por ejemplo deep belief networks es un algoritmo que esta basado en componentes no supervisado llamados restricted Boltzmann machines (RBMs) apilados unos con otros, estos se entrenan de manera secuencial y no suervisada y despues se afinan de manera supervisada 
### Aprendizaje reforzado
- en el aprendizaje reforzado un agente puede observar el ambiente y ejecutar acciones por las que sera recompenzado o penalizado, y en base a esto aprender las mejores estrategias llamadas politicas
- una politica determina cual accion sera tomada en cada situacion
- Con cada iteracion los agentes tendran que actualizar la politica
## Aprendizaje online y batch
- Otro criterio que clasifica si el sistema puede aprender de manera incremental apartir de un flujo entrante de datos
### Aprendizaje batch
- en este tipo de aprendizaje el modelo es entrenado una vez y puesto a produccion
- si los datos cambian, se debe volver a reentrenar el modelo con todos los datos, lo cual es costoso y toma tiempo
### Aprendizaje Online
- En este tipo de aprendizaje el sistema puede recibir flujos constantes de datos ya sea en instancias unicas o multiples (Llamados mini batches)
- El aprendizaje es menos costoso en terminos computacionales
- Puede ser usado para entrenar algoritmos con grandes cantidades de datos que no  caben en memomoria, lo que se llama out-of-core learning
- un parametro importante en este tipo de aprendizaje es el learning rate que representa que tan rapido el algorimo se adapta a los nuevos datos
- un learning rate alto se traduce en que se adaptara rapidamente a nuevos datos pero olvidara los datos antiguos
- Un learning rate bajo el sistema aprendera mas lento pero sera menos suceptible a datos no representativos
- un problema de este tipo de aprendizaje es detectar si datos erroneos estan entrando al sistema, por lo que debe ser monitorizado 
## Instance-based vs Model-based learning
- este tipo de clasificacion determina como el sistema ML generaliza, el sistema debe ser capaz de generalizar ejemplos de datos que nunca ha visto antes, de nada sirve que el sistema tenga un buen performance en los datos de entrenamiento, debe tener un buen performance en datos nuevos
### Instance-based learning
- En este tipo de aprendizaje el sistema aprende de ejemplos y despues generaliza hacia nuevos casos, lo que requiere una medida de similaridad
### Model-based learning
- en este tipo se construye un modelo predictivo a partir de los datos
- Los datos se modelan con el modelo elegido
- despues se debe definir una medida de performance 
- se puede definir una utility function que mide que tan bueno es el modelo o una cost function que mide que tan malo es el modelo # Capitulo 1

## Code example 1-1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data_root = "https://github.com/ageron/data/raw/main/"
life_satisfaction = pd.read_csv(data_root + "lifesat/lifesat.csv")
life_satisfaction
```

### Linear regression (model-based)

```python
from sklearn.linear_model import LinearRegression

x_range = np.linspace(23500, 62500, 100)
x_range_2d = x_range.reshape(-1, 1)
x = life_satisfaction[["GDP per capita (USD)"]].values
y = life_satisfaction[["Life satisfaction"]].values

life_satisfaction.plot(kind="scatter", grid=True, 
					   x="GDP per capita (USD)", y="Life satisfaction")
					   
model = LinearRegression()
model.fit(x, y)

plt.plot(x_range, model.predict(x_range_2d), color="red", label="Prediction")

x_new = [[37655.2]]
prediction = model.predict(x_new)

plt.title("Linear regression")
plt.scatter(x_new, prediction, color="green", marker="^", label=f"Cyprus life satisfactory prediction = {prediction}")
plt.axis([23500, 62500, 4, 9])
plt.legend()
plt.show()
```

### K-nearest neighbor (Instance-based)

```python
from sklearn.neighbors import KNeighborsRegressor

x_range = np.linspace(23500, 62500, 100)
x_range_2d = x_range.reshape(-1, 1)
x = life_satisfaction[["GDP per capita (USD)"]].values
y = life_satisfaction[["Life satisfaction"]].values

life_satisfaction.plot(kind="scatter", grid=True, 
					   x="GDP per capita (USD)", y="Life satisfaction")
					   
model = KNeighborsRegressor(n_neighbors=3)
model.fit(x, y)
plt.plot(x_range, model.predict(x_range_2d), color="red", label="Prediction")

x_new = [[37655.2]]
prediction = model.predict(x_new)

plt.title("K-nearest neighbor")
plt.scatter(x_new, prediction, color="green", marker="^", label=f"Cyprus life satisfactory prediction = {prediction}")
plt.axis([23500, 62500, 4, 9])
plt.legend()
plt.show()
```

## Main challenges of machine learning
### Insufficient quantity of training data
- Incluso para problemas sencillos los algoritmos necesitan una gran cantidad de datos
### Datos de entrenamiento no representativos
- Los datos de entrenamiento deben ser represantativos en funcion del caso de uso que queremos aplicar
- el uso de datos poco representativos dificulta las predicciones precisas
- Cuando el tamao de la muestra de los datos de entrenamiento es muy pequeo puede causar ruido de muestreo (sampling noise), lo que significa que los datos no reflejan fielmente las caracteristicas reales
- Si la muestra es muy grande pero esta mal seleccionada, los datos pueden estar sesgados lo que se conoce como sampling bias
### Baja calidad de los datos
- Si los datos tienen muchos errores, ruido o outliers (datos que se desvian mucho) va a ser mas complicado que el sistema se desempee de buena manera
- Es muy valioso tomarse el tiempo de limpiar el training dataset, ya sea descartar los outliers o arreglar manualmente los errores, o si algunas instancias les falta una caracteristica decidir si descartar esta caracteristica o no
### Carcteristicas irrelevantes
- el sistema necesita caracteristicas relevantes para poder ser entrenado, esto es conocido como feature engineering
    - feature selection, selecciona las caracteristicas mas importantes
    - Feature extraction, combinar caracteristicas existentes para crear una mas relevante
    - crear nuevas caracteristicas reuniendo datos
### Overfitting the training data
- Los algoritmos pueden caer en la sobregeneralizacion, esto se conoce como overfitting, o sea que el algoritmo se desempea bien en el training data pero no es bueno generalizando 
- si los datos tienen mucho ruido es probable que el modelo caiga en esto
- esto pasa cuando el modelo es muy complejo en relacion con la cantidad de ruido en los datos de entrenaiento
- algunas posibles soluciones:
    - reunir mas datos
    - limpiar los datos
    - seleccionar un modelo con menos parametros para reducir su complejidad, esto se conoce como regularizacion
- la cantidad de regularizacion aplicada a un algoritmo de aprendizaje puede se controlado por un hyperparametro definido antes del entrenamento
- un hyperparametro de regularizacion muy alto provocara un modelo casi plano que no puede ser una buena solucion
### Underfitting the training data
- es lo contrario a overfitting, ocurre uando el modelo es demasiado simple como para aprender lo complejo que son los datos 
- para areglar esto
    - seleccionar un modelo con mas parametros 
    - mejorar las caracteristicas
    - reducir el hiperparametro de regularizacion
### Testing and validating
- se puede dividir los datos en datos de entrenamiento de testeso y de entrenamiento
- el error en los nuevos casos se llama generalization error
- si el error de entrenamiento es bajo pero el de generalizacion es alto, significa que el modelo esta overfitting
- 80% entrenamiento y 20% de testeo es lo comun pero esto depende del tamao del dataset
### Hyperparameter tuning and model selection
- al testear el modelo multiples veces puede ocurrir que este se ajusta muy bien a los datos de testeo y el hiperparametro de regularizacion se adapta a ese set en especifico, esto es perjudicial
- una solucion de esto es conocida como holdout vlaidation, en la que se separa una parte del training set para escoger el mejor modelo a usar, este set se conoce como validatiion set o dev set
- se entrena multiples modelos con multiples hiperparametros con el validation set y el mejor despues se entrena con el training set completo
- si el validation set es muy pequelo, las evaluaciones de los modelos seran imprecisas, en cambio si muy grande, los datos de entrenamiento seran muy pequeos
- para solucionar esto se puede aplicar cross-validation usando muchos pequeos validation sets, cada modelo se evaluara con cada uno de los validation set, y el promedio de su desempeo, definira cual es el mejor modelo sin embargo esto retrasara mucho el proceso de entrenamiento
### data mismatch
- los datos de testeo y de validacion deben ser los mas representativos
- si se tiene un problema de data mismatch, en el que los datos no representan la realidad del problema, para saber si es data mismatch o overfitting, se debe separar una parte del data set llamado train-dev set, despues de entrenado el modelo si se desempea bien en el train-dev set esto no sera overfitting, sera data mismath 
## Free launch theorem
- este dice que asumir cosas acerca del data set que tenemos es importante, ya que a priori es imposible definir si un modelo es mejor que otro. tenemos que asumir que nuestros datos funcionaran mejor con un modelo u otro
